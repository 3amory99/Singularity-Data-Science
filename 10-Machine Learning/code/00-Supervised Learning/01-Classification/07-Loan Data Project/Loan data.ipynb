```python
checked_cols = ['int.rate', 'installment', 'log.annual.inc',
       'dti', 'fico', 'days.with.cr.line', 'revol.bal', 'revol.util',
       'inq.last.6mths', 'delinq.2yrs', 'pub.rec']
outliers = ds.structdata.detect_outliers(df, 0, checked_cols)
outliers
```

```python
df[df.index.isin(outliers)]
```

```python
no_outlier_df = df[~df.index.isin(outliers)]
no_outlier_df
```

```python
no_outlier_df = pd.get_dummies(no_outlier_df, columns=['purpose'], drop_first=True)
no_outlier_df
```

```python
df['not.fully.paid'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = no_outlier_df.drop('not.fully.paid', axis=1)
y = no_outlier_df['not.fully.paid']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=22)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=22)
x_train, y_train = sm.fit_sample(x_train, y_train)
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
