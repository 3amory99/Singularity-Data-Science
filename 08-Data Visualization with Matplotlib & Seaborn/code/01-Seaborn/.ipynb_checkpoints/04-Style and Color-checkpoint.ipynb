# Size and Color Palettes

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sns

%matplotlib inline
tips = sns.load_dataset('tips')
```

## 1) Size and Aspect

* 1. change plot size and font scale with sns.set param
* 2. change plot size with height and aspect

### 1. change plot size and font scale with sns.set method

You can use **sns.set()** to change the size and the font scale of most seaborn plots.

```python
sns.set(rc={'figure.figsize': [8, 8]}, font_scale=1.2)
```

```python
sns.countplot(x='sex', data=tips)
```

### 2. change plot size with height and aspect

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: **hight, aspect.** For example:

```python
sns.lmplot(x='total_bill', y='tip', data=tips, height=8, aspect=1.2)
```

or you can use this awesome **cubehelix_palette**

```python
cmap = sns.cubehelix_palette(as_cmap=True, reverse=True, dark=0, light=1)
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(values='passengers', index='month', columns='year')
sns.heatmap(flights, cmap=cmap)
```
